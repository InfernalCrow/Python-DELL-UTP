# ELSE

Es una forma de ejecutar c√≥digo si la condicion if no se cumple

```python
clima = input("""
Como esta el clima?
- soleado
- lluvioso
- nublado
""")

if clima == "soleado":
    print ("El dia esta soleado, podemos ir al parqueüåûüèûÔ∏èüö∂ ")
else:
    print("No conozco ese clima ü§î")
```

```python
edad = 18

if edad >= 18:
    print("Eres mayor de edad!")
else:
    print("Todavia eres menor de edad!")
```

# ELIF

Con esta condici√≥n `elif` agregamos la posibilidad de incluir una condicion si la primera no se cumple.

```python
clima = input("""
Como esta el clima?
- soleado
- lluvioso
- nublado
""")

if clima == "soleado":
    print ("El dia esta soleado, podemos ir al parqueüåûüèûÔ∏èüö∂ ")
elif clima == "lluvioso":
    print ("El dia esta lluvioso, mejor nos quedamos en la casa üè†‚õàÔ∏è")
elif clima == "nublado":
    print("El dia esta nublado, ten cuidado con los rayos ‚ö°")
else:
    print("No conozco ese clima ü§î")
```

1. Escribe un programa en Python para convertir temperaturas a y desde celsius, fahrenheit.

[F√≥rmula : c/5 = f-32/9 [ donde c = temperatura en celsius y f = temperatura en fahrenheit ]

Resultado esperado :
60¬∞C es 140 en Fahrenheit
45¬∞F es 7 en Celsius


```python
print("Programa Para convertir Temperaturas")
print("Ingrese C para convertir de Celsius a fahrenheit")
print("Ingrese F para Convertir de Fahrenheit a Celcius")
condicion=input()

if condicion.lower() == "c":
  
    print ("Ingrese la Temperatura en Celsius")
    tem_c=float(input())
    temperatura_f=float(tem_c*1.8+32)
    print("La Temperatura en Fahrenheit ES: {0}".format(temperatura_f))
    
elif condicion.upper() == "F":
  
    print ("Ingrese la Temperatura en Fahrenheit")
    tem_f=float(input())
    temperatura_c=float(tem_f-32)/1.8
    print("La Temperatura en Celsius ES: {0}".format(int(temperatura_c)))

else:
    print("Error, no existe esa temperatura")
```


2. Escribe un programa en Python para calcular la edad de un perro en a√±os de perro
Nota: Durante los dos primeros a√±os, un a√±o canino equivale a 10,5 a√±os humanos. Despu√©s, cada a√±o canino equivale a 4 a√±os humanos.
Resultado esperado:

Introduzca la edad de un perro en a√±os humanos: 15                                    
La edad del perro en a√±os de perro es 73

```python


anios = int(input("A√±os humanos?: "))

edad_perro = 2*10.5

if anios > 2:
  edad_perro += (anios - 2)*4

print("La edad en a√±os perro es", int(edad_perro))


```
